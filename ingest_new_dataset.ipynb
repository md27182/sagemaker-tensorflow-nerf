## Generate Poses

```python
import numpy as np 

debug = False 

def print_debug(msg):
    if debug:
        print(msg)

intrinsics = np.load(dataset_path + '/calibration_data/cam_mtx_list.npy')
extrinsics = np.load(dataset_path + '/calibration_data/cam_extrinsics.npy')
locations = np.load(dataset_path + '/calibration_data/cam_locations.npy')
print(f'Intrinsics have shape: {intrinsics.shape}')
print(f'Extrinsics have shape: {extrinsics.shape}')
print(f'Locations have shape: {locations.shape}')

# Eventually, we can do everything in a vectorized way, but for now, to make things simple,
# lets break apart each matrix
intrinsics = [intr for intr in intrinsics]
extrinsics = [ext for ext in extrinsics]
locations = [loc for loc in locations]

w = 6000
h = 4000
num_camera_columns = 120

# Cosine and sine of 180 degrees
c = -1 
s = 0
rot_x = np.array([
    [1, 0, 0, 0],
    [0, c, -s, 0],
    [0, s, c, 0],
    [0, 0, 0, 1]
])

bounds_w = 12 # Value is in cm, relative to camera-to-origin distance calculated below

poses_bounds = []
rows = []

for index, (ext, loc) in enumerate(zip(extrinsics, locations)):
    # Figure out which "row" this camera is in: we have to do this because the matrix of
    # camera intrinsics has shape (27, 3, 3), while the other two arrays have shape (3240, ...) - 
    # this is because we only create one intrinsic matrix per "row" of the scanner bed
    row = index // num_camera_columns
    rows.append(row)
    print_debug(f'{index}, {row}')
    intr = intrinsics[row]
    #print(f'Intrs diag: {intr[0,0]}, {intr[1,1]}')
    

    # Apply an additional rotation of 180-degrees about the x-axis (OpenCV -> OpenGL convention)
    #ext = np.matmul(ext, rot_x)
    ext = np.matmul(rot_x, ext)

    # Convert world-to-camera to camera-to-world
    ext = np.linalg.inv(ext)
    print_debug(f'Before \n{ext}')
    
    # Get rid of the last row, which is always(0, 0, 0, 1)
    ext = ext[:3, :]

    # Flip the x and y columns
    ext = ext[:, [1, 0, 2, 3]]

    # Multiply the y column by -1
    ext[:, 0] *= -1
    print_debug(f'After \n{ext}\n')
    
    
    # Append a 5th column to the matrix, which is [h, w, focal]
    focal = (intr[0,0] + intr[1,1]) * 0.5
    last_column = np.array([h, w, focal])
    params = np.c_[ext, last_column]
    print_debug(params)
    

    # Calculate the distance from the camera to the origin using the calibration data
    distance_to_origin = np.linalg.norm(loc)
    print_debug(f'Distance to origin: {distance_to_origin}')

    # The "near" and "far" planes are just set to the distance calculated above +/- a fixed constant
    near = distance_to_origin - bounds_w # np.min(distance_to_origin - bounds_w, loc[1]) # loc[1] is the height of the camera off the stage
    far = 2.5 * distance_to_origin

    # Flatten the 3x5 matrix into a 15-element row vector, append near and far bounds
    params = params.reshape((15,))
    params = np.append(params, [near, far])
    print_debug(f'After reshaping: {params}')
    
    poses_bounds.append(params)

poses_bounds = np.array(poses_bounds)
print(f'All params shape: {poses_bounds.shape}')

rows = np.array(rows)

# Save .npy file
np.save(dataset_path + '/poses_bounds.npy', poses_bounds)
np.save(dataset_path + '/rows.npy', rows)
```

## Calculate which of the source cameras to remove so the spacing is more uniform

```python
import math
import numpy as np 

locations = np.load(dataset_path + '/calibration_data/cam_locations.npy')
num_cols = 120
num_rows = 27

def calculate_avg_theta(row_values):
    sum_theta = 0.0

    for index in range(len(row_values)):
        a = row_values[(index + 0) % len(row_values)] 
        b = row_values[(index + 1) % len(row_values)]

        # These are classes that exist within the framework I was using at the time (basically,
        # 3-vectors with x,y,z coordinates) - you should be able to translate this into numpy
        # code, and calculate the L2 norm to normalize, etc.
        a_dir = a/np.linalg.norm(a)
        b_dir = b/np.linalg.norm(b)

        theta = math.acos(np.dot(a_dir, b_dir))
        sum_theta += theta

    # Divide by the number of pairs
    sum_theta /= len(row_values)
    sum_theta *= (180.0 / math.pi)

    return sum_theta

def find_optimal_spacing(row_values, desired_theta=3.0):
    closest_theta = calculate_avg_theta(row_values)
    closest_spacing = 1
    starting_theta = closest_theta

    keep_every = list(range(2, 24))

    assert len(row_values) == num_cols

    for entry in keep_every:

        row_copy = [row_values[index] for index in range(len(row_values)) if index % entry == 0]
        theta = calculate_avg_theta(row_copy)

        if abs(theta - desired_theta) < abs(closest_theta - desired_theta):
            closest_theta = theta 
            closest_spacing = entry 

    print(f'\tOptimal spacing found with angle of separation {closest_theta}: use every {closest_spacing} entries')

    return closest_spacing

# These are the things you ultimately need from this script - the new,
# "optimized" camera locations (and their indices, w.r.t. the 3240 
# original indices)
usable_indices = []
optimized_locations = []

for row_index in range(num_rows):
    # Grab the camera locations corresponding to this row
    row_values = locations[row_index * num_cols : (row_index + 1) * num_cols]
    theta = calculate_avg_theta(row_values)

    print(f'For row {row_index}, avg theta between neighbors: {theta}')
    closest_spacing = find_optimal_spacing(row_values)

#     optimized = [row_values[index] for index in range(len(row_values)) if index % closest_spacing == 0]

#     for loc in optimized:
#         optimized_locations.append(loc)

    usable_indices += [(row_index * num_cols + index) for index in range(len(row_values)) if index % closest_spacing == 0]
    
# Skip every x indices to get the total number of images closer to TARGET_NUM_IMAGES
if TARGET_NUM_IMAGES is not None:
    use_every = len(usable_indices) // TARGET_NUM_IMAGES + 1
else:
    use_every = 1
    
usable_indices = usable_indices[::use_every]
    
print('final number of images is ', len(usable_indices))


```

## Copy the subset of images and poses into their final location
May take upwards of 5-10 minutes depending on the number of images to copy.

```python
import os
import glob 
import numpy as np
from shutil import copyfile

dest_path = '/home/ec2-user/SageMaker/lbx-nerf/data/lbx'
dataset = 'sl2'
resolutions = ['images'] # e.g. ['images', 'images_8']

sparsify = False

for res in resolutions:
    os.makedirs(f'{dest_path}/{dataset}/{res}', exist_ok=True)

# Grab paths to all of the relevant images
images = []
images_folder = f'{dataset_path}/white/*.jpg'
for file in sorted(glob.glob(images_folder)):
    images.append(file)
number_of_images = len(images)

# Load the poses (matrices)
poses_file = f'{dataset_path}/poses_bounds.npy'
poses = np.load(poses_file)
number_of_poses = len(poses)

# Load the rows
rows_file = f'{dataset_path}/rows.npy'
rows = np.load(rows_file)

assert number_of_images == number_of_poses
assert number_of_images == len(rows)
print(f'Processing {number_of_images} images and {number_of_poses} poses')
print(f'Poses array has shape: {poses.shape}')

# subset the images and poses
if sparsify:
    images_pruned = [images[i] for i in usable_indices]
    poses_pruned = poses[usable_indices]
    rows_pruned = rows[usable_indices]
else:
    images_pruned = images
    poses_pruned = poses
    rows_pruned = rows
print(f'Pruned images array has length: {len(images_pruned)}')
print(f'Pruned poses array has shape: {poses_pruned.shape}')
assert len(images_pruned) == poses_pruned.shape[0]

poses_pruned_file = f'{dest_path}/{dataset}/poses_bounds.npy'
np.save(poses_pruned_file, poses_pruned)

rows_pruned_file = f'{dest_path}/{dataset}/rows.npy'
np.save(rows_pruned_file, rows_pruned)

# copy intrinsics file into destination directory
dest_intrinsics_path = f'{dest_path}/{dataset}/cam_mtx_list.npy'
copyfile(dataset_path + '/calibration_data/cam_mtx_list.npy', dest_intrinsics_path)
```

```python
for src in images_pruned:
    # Extract the file name from the full path, i.e. /clamp/images_8/my_image_0000.jpg -> my_image_0000.jpg
    basename = os.path.basename(src)
    for res in resolutions:
        dst = f'{dest_path}/{dataset}/{res}/{basename}'

        # File copy
        copyfile(src, dst)
```
